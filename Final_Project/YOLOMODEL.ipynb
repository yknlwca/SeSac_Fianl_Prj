```python
import glob
import os
import cv2
import xml.etree.ElementTree as ET


xml_directory = 'C:/Users/user/Desktop/SeSaC/Final_Project/Abnormal/Theft_xml' 
frames_dir = 'C:/Users/user/Desktop/SeSaC/Final_Project/Abnormal/jin'  

if not os.path.exists(frames_dir):
    os.makedirs(frames_dir)

def extract_theft_frames(xml_file_path):
    tree = ET.parse(xml_file_path)
    root = tree.getroot()

    theft_start_frame = None
    theft_end_frame = None

    for track in root.findall('track'):
        if track.get('label') == 'theft_start':
            theft_start_frame = int(track.find('box').get('frame'))
        elif track.get('label') == 'theft_end':
            theft_end_frame = int(track.find('box').get('frame'))

    return theft_start_frame, theft_end_frame

def process_video(file_path, start_frame, end_frame, frame_dir, file_index):
    cap = cv2.VideoCapture(file_path)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(f'{frame_dir}/theft_cut_{file_index}.mp4', fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

    frame_num = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if start_frame <= frame_num <= end_frame:
            out.write(frame)
        frame_num += 1

    cap.release()
    out.release()
    cv2.destroyAllWindows()

file_index = 1

for xml_file in glob.glob(os.path.join(xml_directory, '*.xml')):
    theft_start_frame, theft_end_frame = extract_theft_frames(xml_file)
    if theft_start_frame is not None and theft_end_frame is not None:
        video_file_name = os.path.basename(xml_file).replace('.xml', '.mp4')
        video_path = os.path.join('C:/Users/user/Desktop/SeSaC/Final_Project/Abnormal/Theft_Video', video_file_name)
        process_video(video_path, theft_start_frame, theft_end_frame, frames_dir, file_index)
        file_index += 1
```

```python
import cv2
import os

video_folder = 'C:/Users/user/Desktop/SeSaC/Final_Project/Abnormal/jin'

frames_dir = 'C:/Users/user/Desktop/SeSaC/Final_Project/Abnormal/Theft_frame'

video_extension = '.mp4'

for video_file in os.listdir(video_folder):
    if video_file.endswith(video_extension):
        video_path = os.path.join(video_folder, video_file)
        cap = cv2.VideoCapture(video_path)

        video_frames_dir = os.path.join(frames_dir, os.path.splitext(video_file)[0])
        if not os.path.exists(video_frames_dir):
            os.makedirs(video_frames_dir)

        frame_num = 0
        while cap.isOpened():
            ret, frame = cap.read()

            if not ret:
                break

            frame_path = os.path.join(video_frames_dir, f'frame_{frame_num}.jpg')

            if frame is not None:
                cv2.imwrite(frame_path, frame)
                frame_num += 1

        cap.release()
```

```python
from ultralytics import YOLO
import glob
import os


model = YOLO('yolov8n-pose.pt')

base_directory = 'C:/Users/user/Desktop/SeSaC/Final_Project/Abnormal/Theft_frame'

for folder_name in os.listdir(base_directory):
    folder_path = os.path.join(base_directory, folder_name)
    if not os.path.isdir(folder_path):
        continue

    for image_file in glob.glob(f'{folder_path}/*.jpg'):
        results = model(source=image_file)
        
        for result in results:
            output_file_path = f'{folder_path}/{os.path.basename(image_file).replace(".jpg", ".txt")}'
            with open(output_file_path, 'w') as txt_file:
                if len(result.boxes.xyxy) == 0:
                    continue

                # 바운딩 박스 및 키포인트 처리
                for box in result.boxes.xyxy:
                    class_id = 2
                    x1, y1, x2, y2 = [coord.item() for coord in box[:4]]
                    x1, y1, x2, y2 = x1 / 1920, y1 / 1080, x2 / 1920, y2 / 1080

                    txt_file.write(f'{class_id} {x1:.6f} {y1:.6f} {x2:.6f} {y2:.6f} ')

                for keypoint in result.keypoints.xyn:
                    x, y = keypoint[0].numpy(), keypoint[1].numpy()  
                    txt_file.write(f'{x[0]:.6f} {y[0]:.6f} ')  
```
